## ==============================================================================
## MODULE 1: HARDWARE THREAD AUDITING & BULK COHORT INGESTION
## ==============================================================================
**Mechanics:** High-Performance Computing (HPC) requires identifying available processing lanes. We audit our CPU thread count programmatically before spawning parallel task pools.

```python
print("-------------------------------------------------------------------------")
print("[STAGE 1]: Profiling Available CPU Core Threads & Hardware Topography...")
print("-------------------------------------------------------------------------")

import os
import sys
import time
import json
import numpy as np
import multiprocessing as mp
import matplotlib.pyplot as plt

available_cores = mp.cpu_count()
print(f"[CI AUDIT] Multi-Threaded CPU Engine Online.")
print(f"[CI AUDIT] Total Hardware Processing Lanes Available: {available_cores} Cores")

print("\n-------------------------------------------------------------------------")
print("[STAGE 2]: Generating Multi-Subject Asynchronous Population Cohort...")
print("-------------------------------------------------------------------------")

def generate_bulk_cohort_dataset(num_subjects=50):
    """
    Simulates a high-throughput programmatic stream from an AWS S3 neuro-bucket.
    Populates an indexed population data bank containing 50 distinct records.
    """
    np.random.seed(101) 
    cohort_dict = {}
    time_points = 400
    channels = 16
    
    print(f"[INGESTION]: Instantiating parallel connection channels to S3 endpoint...")
    for sub_idx in range(1, num_subjects + 1):
        subj_id = f"SUBJ-{sub_idx:03d}"
        
        raw_signal = np.random.normal(0.0, 1.2, (channels, time_points))
        t = np.linspace(0, 8, time_points)
        
        individual_shift = np.random.uniform(1.0, 4.0)
        for i in range(4):
            raw_signal[i, :] += np.sin(t) * individual_shift
            
        cohort_dict[subj_id] = {
            "subject_id": subj_id,
            "sampling_rate": 100,
            "raw_voltage": raw_signal.tolist(),
            "true_kinematics": (np.sin(t) * (individual_shift * 0.5)).tolist()
        }
        
    print(f"[SUCCESS]: Streamed and allocated {len(cohort_dict)} subject entries in system memory.")
    return cohort_dict

population_payload = generate_bulk_cohort_dataset(num_subjects=50)
```

## ==============================================================================
## MODULE 2: ESTABLISHING THE BASELINE SEQUENTIAL COMPUTE WALL
## ==============================================================================
**Mechanics:** To demonstrate an infrastructure speedup metric, we implement a baseline tracking function that processes the cohort sequentially using a single CPU thread.

```python
def single_subject_pipeline_unit(subject_data_package):
    """
    Executes pre-processing and intentional decoding on an isolated subject array.
    This encapsulates the exact core workflow logic derived during Colab 1.
    """
    raw_voltage = np.array(subject_data_package["raw_voltage"])
    true_kinematics = np.array(subject_data_package["true_kinematics"])
    
    feat_x = np.mean(raw_voltage[0:4, :], axis=0)
    
    kernel = np.ones(3) / 3.0
    predicted_kinematics = np.convolve(feat_x * 0.45, kernel, mode='same')
    
    mse_score = float(np.mean((true_kinematics - predicted_kinematics) ** 2))
    time.sleep(0.015) 
    
    return subject_data_package["subject_id"], mse_score

print("-------------------------------------------------------------------------")
print("[BENCHMARKING]: Executing Cohort Analysis Sequentially on Core 0...")
print("-------------------------------------------------------------------------")

start_seq_time = time.time()
sequential_results = {}

for s_id, s_data in population_payload.items():
    res_id, score = single_subject_pipeline_unit(s_data)
    sequential_results[res_id] = score

total_seq_duration = time.time() - start_seq_time
print(f"[RESULT]: Sequential processing loop took: {total_seq_duration:.4f} seconds.")
```

## ==============================================================================
## MODULE 3: THE PARALLEL WORKPOOL ENGINE (MULTIPROCESSING ACCELERATION)
## ==============================================================================
### 💡 STUDENT AI PROMPT CHALLENGE:
Copy and paste the exact text inside the block quotation below into your AI assistant:

> "Write a high-performance parallel processing function named 'execute_parallel_hpc_pipeline(cohort_data, core_count)' that accelerates our multi-subject analysis across all available hardware processing lanes. The function must satisfy these structural parameters:
> 1. Initialize a Python multiprocessing worker pool using 'mp.Pool(processes=core_count)'.
> 2. Extract the individual subject dictionaries from the parent 'cohort_data' master matrix and place them into a flat python list format to prepare a parallel map payload.
> 3. Distribute the workload task queue asynchronously across the multi-core allocation using the pool's '.map()' execution framework, targeted at our atomic 'single_subject_pipeline_unit' worker function.
> 4. Cleanly close out the active worker threads and join the lanes back together.
> 5. Return the mapped list of completed result structures."

```python
# 🧠 STUDENT ZONE: PASTE THE GENERATED PARALLEL SUB-SYSTEM ENGINE AND CO-AUTHORED CODES HERE:
def execute_parallel_hpc_pipeline(cohort_data, core_count):
    subject_tasks_list = [data for s_id, data in cohort_data.items()]
    
    with mp.Pool(processes=core_count) as hpc_pool:
        mapped_results = hpc_pool.map(single_subject_pipeline_unit, subject_tasks_list)
        
    return mapped_results

hpc_allocation = available_cores

print("-------------------------------------------------------------------------")
print(f"[HPC ACTION]: Spinning up Parallel Workpool Mapping Across {hpc_allocation} Cores...")
print("-------------------------------------------------------------------------")

start_par_time = time.time()
parallel_raw_output = execute_parallel_hpc_pipeline(population_payload, hpc_allocation)
total_par_duration = time.time() - start_par_time

parallel_results = {item[0]: item[1] for item in parallel_raw_output}
print(f"[RESULT]: Multi-core parallel mapping loop completed in: {total_par_duration:.4f} seconds.")
```

## MODULE 4: POPULATION TELEMETRY SHIELD & SPEEDUP METRIC RENDERING
## ==============================================================================
### 💡 STUDENT AI PROMPT CHALLENGE:
Copy and paste the exact text inside the block quotation below into your AI assistant:

> "Write a comprehensive dual-axis benchmarking graph function named 'render_scaling_telemetry(seq_time, par_time, results_dict)'. The layout frame must feature an exact 2-column horizontal orientation viewport (1 row, 2 columns):
> 1. Panel 1 (Left Population Analytics Chart): Plot a histogram showing the distribution of Mean Squared Error values across all 50 subjects to establish cohort consistency bounds.
> 2. Panel 2 (Right Cyberinfrastructure Performance Bar Chart): Draw a comparative bar plot displaying 'Sequential (Single-Core)' processing time directly against 'Parallel (Multi-Core)' processing time. Calculate the empirical Speedup Factor Scalar (Speedup = Sequential_Time / Parallel_Time) and print it as a large status card title text inside this panel frame.
> Apply clear layout styling using tight borders, data grids, and distinct colors."

```python
# 🧠 STUDENT ZONE: PASTE THE GENERATED GRAPHICS INTERFACE DESIGN DIRECTLY HERE:
def render_scaling_telemetry(seq_time, par_time, results_dict):
    mse_scores = list(results_dict.values())
    speedup_factor = seq_time / par_time
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    ax1.hist(mse_scores, bins=10, color='#1f77b4', edgecolor='black', alpha=0.8)
    ax1.axvline(np.mean(mse_scores), color='red', linestyle='--', linewidth=2, label=f"Cohort Mean: {np.mean(mse_scores):.4f}")
    ax1.set_title("Population Distribution Dashboard — Accuracy Verification Layer", fontsize=11, fontweight='bold')
    ax1.set_xlabel("Mean Squared Error Score Bounds", fontsize=10)
    ax1.set_ylabel("Subject Track Frequency Count", fontsize=10)
    ax1.grid(True, linestyle=':', alpha=0.5)
    ax1.legend()
    
    execution_modes = ['Sequential\n(Core 0 Only)', f'Parallel\n({available_cores} Active Lanes)']
    execution_durations = [seq_time, par_time]
    colors = ['#ff7f0e', '#2ca02c']
    
    bars = ax2.bar(execution_modes, execution_durations, color=colors, edgecolor='black', width=0.5)
    ax2.set_ylabel("Total Operational Execution Clock Latency (Seconds)", fontsize=10)
    ax2.set_title(f"HPC Infrastructure Accelerating Shield \nEmpirical Speedup Factor: {speedup_factor:.2fx} Faster", fontsize=11, color='#155724' if speedup_factor > 1.2 else '#721c24', fontweight='bold')
    ax2.grid(True, linestyle=':', axis='y', alpha=0.5)
    
    for bar in bars:
        height = bar.get_height()
        ax2.annotate(f'{height:.3f}s',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom', weight='bold')
        
    plt.tight_layout()
    plt.show()
    return speedup_factor

empirical_speedup = render_scaling_telemetry(total_seq_duration, total_par_duration, parallel_results)
```

## ==============================================================================
## MODULE 5: WORKFLOW PACKAGE ENVIRONMENT SNAPSHOT (FAIR CI COMPLIANCE)
## ==============================================================================
**Mechanics:** True reproducibility means documenting not just data origins, but capturing the exact software environment snapshot indices (`pip freeze`) within the virtual runtime container.

```python
print("-------------------------------------------------------------------------")
print("[FAIR COMPLIANCE]: Serializing Environment Hashes & Compute Run Records...")
print("-------------------------------------------------------------------------\n")

import subprocess

def export_global_hpc_provenance(cores_used, seq_time, par_time, speedup):
    try:
        pip_snapshot = subprocess.check_output([sys.executable, "-m", "pip", "freeze"]).decode().split('\n')
        core_packages = [pkg for pkg in pip_snapshot if any(x in pkg for x in ['numpy', 'matplotlib', 'scipy'])]
    except:
        core_packages = ["Standard NumPy Framework Native Context Extraction Only Available"]

    global_provenance = {
        "hpc_environment_specifications": {
            "allocated_cpu_workers": cores_used,
            "python_runtime_binary_version": sys.version.split()[0],
            "monitored_environmental_hashes": core_packages
        },
        "performance_benchmarking_metrics": {
            "baseline_sequential_seconds": seq_time,
            "accelerated_parallel_seconds": par_time,
            "calculated_speedup_coefficient": speedup
        },
        "institutional_compliance_verdict": "CERTIFIED REPRODUCIBLE" if speedup > 1.0 else "UNACCELERATED RUN WARNING"
    }
    
    output_filename = "global_cohort_provenance_manifest.json"
    with open(output_filename, 'w') as json_file:
        json.dump(global_provenance, json_file, indent=4)
        
    print(f"[METADATA SUCCESS]: Universal scaling tracking file saved to disk as: '{output_filename}'")
    print(json.dumps(global_provenance, indent=2))

export_global_hpc_provenance(hpc_allocation, total_seq_duration, total_par_duration, empirical_speedup)
```
