```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

## Walkthrough A: Classification

```python
attention_ml=pd.DataFrame({'sleep':[4,5,6,7,8,9,5,6,7,8,4,5],'exercise':[0,1,2,3,4,5,1,2,3,4,0,1],'focus':[0,0,0,1,1,1,0,0,1,1,0,0]})
attention_ml.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
X=attention_ml[['sleep','exercise']]
y=attention_ml['focus']
Xtr,Xte,ytr,yte=train_test_split(X,y,test_size=0.25,random_state=42)
model=DecisionTreeClassifier(max_depth=3).fit(Xtr,ytr)
print('Accuracy:',model.score(Xte,yte))
```

Reflection: What features helped make predictions?

## Walkthrough B: Regression

```python
sleep_reg=pd.DataFrame({'sleep_hours':[4,5,6,7,8,9,5,6,7,8],'attention':[48,55,64,73,80,88,58,67,75,84]})
```

```python
from sklearn.linear_model import LinearRegression
X=sleep_reg[['sleep_hours']]
y=sleep_reg['attention']
model=LinearRegression().fit(X,y)
print('R2:',model.score(X,y))
```

Does prediction imply causation? Why not?

## Clustering Activity

```python
from sklearn.cluster import KMeans
behavior=pd.DataFrame({'sleep':[4,5,6,8,9,4,5,6,8,9],'stress':[8,8,7,4,3,9,8,6,5,3]})
behavior['cluster']=KMeans(n_clusters=3,random_state=42).fit_predict(behavior[['sleep','stress']])
behavior.head()
```
